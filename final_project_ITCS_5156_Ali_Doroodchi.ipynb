```python
T = df['RP']
X = df.drop('RP', axis=1)
N = len(df)

from sklearn.model_selection import train_test_split
X_train, X_test, t_train, t_test = train_test_split(X, T, test_size=0.2, random_state=0)
print(X.shape, X_train.shape, X_test.shape, t_train.shape, t_test.shape)
```

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
np.random.seed(0)

imputer = SimpleImputer(strategy='mean')
X_train_imputed = imputer.fit_transform(X_train)
X_test_imputed = imputer.transform(X_test)

model = LinearRegression()

model.fit(X_train_imputed, t_train)

train_score = model.score(X_train_imputed, t_train)
test_score = model.score(X_test_imputed, t_test)

print("Train score: ", train_score)
print("Test score: ", test_score)
```

```python
y = model.predict(X_test_imputed)

plt.scatter(t_test, y)
plt.plot([-3,3], [-3,3], 'r--')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title('Linear Regression: Actual vs. Predicted')
plt.show()
```

```python
# Multiple Layer Perceptron
from sklearn.model_selection import train_test_split
X = df.drop(columns=['RP'])
t = df['RP']
X_train, X_test, t_train, t_test = train_test_split(X, t, test_size=0.2, random_state=42)

print("Train data shape: {}".format(X_train.shape))
print("Train target shape: {}".format(t_train.shape))
print("Test data shape: {}".format(X_test.shape))
print("Test target shape: {}".format(t_test.shape))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPRegressor
from sklearn.preprocessing import StandardScaler
print("Data types in X_train:\n", X_train.dtypes)
print("\nData types in X_test:\n", X_test.dtypes)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

mlp_reg = MLPRegressor(hidden_layer_sizes=(50, 50), max_iter=500, random_state=42)

mlp_reg.fit(X_train_scaled, t_train)

train_score = mlp_reg.score(X_train_scaled, t_train)
test_score = mlp_reg.score(X_test_scaled, t_test)

print("Train Accuracy: {:.2f}, Test Accuracy: {:.2f}".format(train_score, test_score))
```

```python
import matplotlib.pyplot as plt
import numpy as np

diag_line = np.linspace(min(t_test.min(), y.min()), max(t_test.max(), y.max()), 100)

y = mlp_reg.predict(X_test_scaled)

plt.scatter(t_test, y)
plt.plot(diag_line, diag_line, color='red', linestyle='--')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Multi Layer Perceptron: Actual vs. Predicted')
plt.show()
```
